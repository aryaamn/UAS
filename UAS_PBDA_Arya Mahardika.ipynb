```python
print('mulai preprocesing data')
print()
print('Case Folding')
case_folding = data.apply(lambda x: x.astype(str).str.lower())
print(case_folding)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
#Stemming
print('Stemming')
factory = StemmerFactory()
stemmer = factory.create_stemmer()
count = 0
x = 300

while count < x:
#     print(count)
    stem_text = stemmer.stem(case_folding['text'][count])
#     print(stem_text)
    case_folding['text'][count] = stem_text
    count += 1

print(case_folding)
```

```python
#Tokenize
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

print("Tokenize")
count = 0
x = 300
while count < x:
    tokens = word_tokenize(stemmer.stem(case_folding['text'][count]))
    case_folding['text'][count] = tokens
    count += 1
print(case_folding)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
#Stopwords
print("Stopwords")
count = 0
x = 300
stopword=set(stopwords.words('indonesian'))
removed=[]
while count < x:
    for t in case_folding['text'][count]:
        if t not in stopword:
            removed.append(t)
#         else :
#            case_folding['text'][count].remove(t)
    count += 1
    
print('Stopword Removal')
print("Removed Item = ",removed)
```
